```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'])
df['transaction_time'].dtype
```

```python
df['transaction_time'].iloc[0]
type(df['transaction_time'].iloc[0])
```

```python
df['hour'] = df['transaction_time'].apply(lambda t: t.hour)
```

```python
df[['transaction_time', 'hour']].head()
```

```python
hourly_revenue = df.groupby('hour')['revenue'].mean()
```

```python
hourly_revenue_sorted  = hourly_revenue.sort_values(ascending = True)
```
